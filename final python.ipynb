As we can see that the dataset has outliers, hence we will treat them. We have created a user definded function for finding the lower and upper range for a variable so that outlier can be treated.

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    
```

```python
lrprice,urprice=remove_outlier(df['price'])
df['price']=np.where(df['price']>urprice,urprice,df['price'])
df['price']=np.where(df['price']<lrprice,lrprice,df['price'])


lrliving_measure,urliving_measure=remove_outlier(df['living_measure'])
df['living_measure']=np.where(df['living_measure']>urliving_measure,urliving_measure,df['living_measure'])
df['living_measure']=np.where(df['living_measure']<lrliving_measure,lrliving_measure,df['living_measure'])

lrlot_measure,urlot_measure=remove_outlier(df['lot_measure'])
df['lot_measure']=np.where(df['lot_measure']>urlot_measure,urlot_measure,df['lot_measure'])
df['lot_measure']=np.where(df['lot_measure']<lrlot_measure,lrlot_measure,df['lot_measure'])

lrceil_measure,urceil_measure=remove_outlier(df['ceil_measure'])
df['ceil_measure']=np.where(df['ceil_measure']>urceil_measure,urceil_measure,df['ceil_measure'])
df['ceil_measure']=np.where(df['ceil_measure']<lrceil_measure,lrceil_measure,df['ceil_measure'])

lrliving_measure15,urliving_measure15=remove_outlier(df['living_measure15'])
df['living_measure15']=np.where(df['living_measure15']>urliving_measure15,urliving_measure15,df['living_measure15'])
df['living_measure15']=np.where(df['living_measure15']<lrliving_measure15,lrliving_measure15,df['living_measure15'])

lrlot_measure15,urlot_measure15=remove_outlier(df['lot_measure15'])
df['lot_measure15']=np.where(df['lot_measure15']>urlot_measure15,urlot_measure15,df['lot_measure15'])
df['lot_measure15']=np.where(df['lot_measure15']<lrlot_measure15,lrlot_measure15,df['lot_measure15'])

lrtotal_area,urtotal_area=remove_outlier(df['total_area'])
df['total_area']=np.where(df['total_area']>urtotal_area,urtotal_area,df['total_area'])
df['total_area']=np.where(df['total_area']<lrtotal_area,lrtotal_area,df['total_area'])
```

```python
# Copy all the predictor variables into X dataframe
X = df2.drop(['price', 'Clus_kmeans', 'sil_width'], axis=1)
# Copy target into the y dataframe. 
y = df2['price']
```

```python
# Split X and y into training and test set in 70:30 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

### Linear Regression using statsmodels

```python
# concatenate X and y into a single dataframe
data_train = pd.concat([X_train,y_train],axis=1)
data_test = pd.concat([X_test,y_test],axis=1)
```

```python
data_train.columns
```

```python
expr= 'price ~ room_bed + room_bath + living_measure + lot_measure +ceil + coast + sight + condition  + quality + ceil_measure + basement + yr_renovated + living_measure15 + lot_measure15 + furnished + total_area + house_age'
```

```python
import statsmodels.formula.api as smf
lm1 = smf.ols(formula=expr,data=data_train).fit()
print(lm1.summary())
```

```python
# Calculate MSE for train data
mse = np.mean((lm1.predict(data_train.drop('price',axis=1))-data_train['price'])**2)
```

```python
#Root Mean Squared Error - RMSE for train data
np.sqrt(mse)
```

```python
# Calculate MSE for test data
mse1 = np.mean((lm1.predict(data_test.drop('price',axis=1))-data_test['price'])**2)
```

```python
#Root Mean Squared Error - RMSE for test data
np.sqrt(mse1)
```

Here we can see that the RMSE of the test data is nearly similar to the RMSE of the train data, hence the model is generalizing well to new data. This means that the model is not just memorizing the training data, but it is also learning to make predictions on new data.

### Interpretation
The final Linear Regression equation is

price = b0 + b1 * room_bed + b2 * room_bath + b3 * living_measure + b4 * lot_measure + b5 * ceil + b6 * coast + b7 * sight + b8 * condition + b9 * quality + b10 * ceil_measure + b11 * basement + b12 * yr_renovated + b13 * living_measure15 + b14 * lot_measure15 + b15 * furnished + b16 * total_area + b17 * house_age

price = (-627068.55) * Intercept + (-18563.44) * room_bed + (25274.64) * room_bath + (87.83) * living_measure + (-9.23) * lot_measure + (34268.52) * ceil + (111546.04) * coast + (24172.66) * sight + (21914.01) * condition + (81771.76) * quality + (-8.09) * ceil_measure + (22060.26) * basement + (24015.18) * yr_renovated + (59.5) * living_measure15 + (-2.79) * lot_measure15 + (60748.6) * furnished + (7.02) * total_area + (2585.17) * house_age

When number of bathrooms increases by 1 unit, price of the house increases by 25274 units, keeping all other predictors constant. Similarly, when no. (levels) of floors increases by 1 unit, house price increases by 34268 units, keeping all other predictors constant, and so on.

There are also some negative co-efficient values, for instance, lot_measure has its corresponding co-efficient as -9.23. This implies, when the lot_measure increases by 1 unit, the house price decreases by 9.23 units, keeping all other predictors constant.
